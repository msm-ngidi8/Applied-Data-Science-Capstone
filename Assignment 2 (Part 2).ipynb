```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Load the data

Neighbourhoods_df = pd.read_csv('Part_1.csv')
```

```python
# Display the first 5 rows of the Neighbourhood

Neighbourhoods_df.head()
```

```python
# Load the data and rename the columns 

Geospatial_Data_df = pd.read_csv('Geospatial_Coordinates.csv')

Geospatial_Data_df.rename(columns={'Postal Code':'PostalCode'}, inplace=True)
```

```python
# Display the first 5 rows of Geospatial Coordinates

Geospatial_Data_df.head()
```

```python
# Add the Geospatial coordinates to the previous Dataframe

Neighbourhoods_with_location = pd.merge(Neighbourhoods_df, Geospatial_Data_df, how='inner', left_on='PostalCode', right_on='PostalCode')
```

```python
# Display all the rows of the dataset

Neighbourhoods_with_location
```

```python
Neighbourhoods_with_location.to_csv(r'Part_2.csv')
```
